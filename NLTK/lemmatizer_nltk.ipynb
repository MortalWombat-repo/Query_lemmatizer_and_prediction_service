```python
import pandas as pd
from kategorije import reset_lozinke, cijena_paketa, problem_prijave, ostalo
import string
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from IPython.display import display
```

```python
'''reset_lozinke DataFrame'''
reset_lozinke_labels = (
    ["reset_lozinke"] * len(reset_lozinke))

df_reset_lozinke = pd.DataFrame({
    "text": reset_lozinke,
    "label": reset_lozinke_labels
})

'''problem_prijave DataFrame'''
problem_prijave_labels = (
    ["problem_prijave"] * len(problem_prijave))


df_problem_prijave = pd.DataFrame({
    "text": problem_prijave,
    "label": problem_prijave_labels
})

'''cijena_paketa DataFrame'''
cijena_paketa_labels = (
    ["cijena_paketa"] * len(cijena_paketa))


df_cijena_paketa = pd.DataFrame({
    "text": cijena_paketa,
    "label": cijena_paketa_labels
})

'''ostalo DataFrame'''
ostalo_labels = (
    ["ostalo"] * len(cijena_paketa))


df_ostalo = pd.DataFrame({
    "text": ostalo,
    "label": ostalo_labels
})

display(df_reset_lozinke)
display(df_problem_prijave)
display(df_cijena_paketa)
display(df_ostalo)
```

```python
# clean text of punctuation
def clean_text(text):
    text = text.lower()
    text = ''.join([char for char in text if char not in string.punctuation])
    return text

# applying the cleaning function to a dataset
df_reset_lozinke['cleaned_text'] = df_reset_lozinke['text'].apply(lambda x: clean_text(x))
df_problem_prijave['cleaned_text'] = df_problem_prijave['text'].apply(lambda x: clean_text(x))
df_cijena_paketa['cleaned_text'] = df_cijena_paketa['text'].apply(lambda x: clean_text(x))
df_ostalo['cleaned_text'] = df_ostalo['text'].apply(lambda x: clean_text(x))


display(df_reset_lozinke)
display(df_problem_prijave)
display(df_cijena_paketa)
display(df_ostalo)
```

```python
''' NLTK in wordnet only in English, yet has best lemmatization'''
# NLTK data for lemmatization
nltk.download('wordnet')

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Function to apply stemming
def stem_text(text):
    return ' '.join([stemmer.stem(word) for word in text.split()])

# Function to apply lemmatization
def lemmatize_text(text):
    return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])

# Applying stemming and lemmatization
#df['stemmed_text'] = df['cleaned_text'].apply(lambda x: stem_text(x))
df_reset_lozinke['lemmatized_text'] = df_reset_lozinke['cleaned_text'].apply(lambda x: lemmatize_text(x))
df_problem_prijave['lemmatized_text'] = df_problem_prijave['cleaned_text'].apply(lambda x: lemmatize_text(x))
df_cijena_paketa['lemmatized_text'] = df_cijena_paketa['cleaned_text'].apply(lambda x: lemmatize_text(x))
df_ostalo['lemmatized_text'] = df_ostalo['cleaned_text'].apply(lambda x: lemmatize_text(x))


display(df_reset_lozinke)
display(df_problem_prijave)
display(df_cijena_paketa)
display(df_ostalo)
```

```python
# Assuming you already have df_reset_lozinke
reset_lozinke_lemma = df_reset_lozinke['lemmatized_text'].tolist()
problem_prijave_lemma = df_problem_prijave['lemmatized_text'].tolist()
cijena_paketa_lemma = df_cijena_paketa['lemmatized_text'].tolist()
ostalo_lemma = df_ostalo['lemmatized_text'].tolist()

# Write to a new Python file
with open("kategorije_lemmatized_nltk.py", "w", encoding="utf-8") as f:
    f.write("reset_lozinke = [\n")
    for text in reset_lozinke_lemma:
        escaped = text.replace("\\", "\\\\").replace('"', '\\"')  # escape special chars
        f.write(f'    "{escaped}",\n')
    f.write("]\n")

    f.write("problem_prijave = [\n")
    for text in problem_prijave_lemma:
        escaped = text.replace("\\", "\\\\").replace('"', '\\"')  # escape special chars
        f.write(f'    "{escaped}",\n')
    f.write("]\n")

    f.write("cijena_paketa = [\n")
    for text in cijena_paketa_lemma:
        escaped = text.replace("\\", "\\\\").replace('"', '\\"')  # escape special chars
        f.write(f'    "{escaped}",\n')
    f.write("]\n")

    f.write("ostalo = [\n")
    for text in ostalo_lemma:
        escaped = text.replace("\\", "\\\\").replace('"', '\\"')  # escape special chars
        f.write(f'    "{escaped}",\n')
    f.write("]\n")
```
